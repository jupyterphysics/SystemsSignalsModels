# Problem 19.4 - Continuous correlation

Import the packages

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

Define the functions and construc a sequence of data

```python
num = 500
tstart = 0.0
tend = 10.0
tt = np.linspace(tstart, tend, num)
sig1 = np.sin(tt)
sig2 = np.cos(2*tt)
```

```python
plt.plot(tt,sig1,color='red',label='sin(t)')
plt.plot(tt,sig2,color='blue',label='cos(2t)')
plt.xlabel('$t [s]$')
plt.ylabel('$signals$')
plt.grid(True)
plt.legend()
plt.title('presentation of the two signals')
plt.show()
```

Calculation of the correlations

```python
c12 = np.correlate(sig1,sig2,"full")
c11 = np.correlate(sig1,sig1,"full")
c22 = np.correlate(sig2,sig2,"full")
```

```python
ttc = np.linspace(-tend, tend, 2*num-1)
plt.plot(ttc,c12,color='red',label="cross-correlation")
plt.plot(ttc,c11,color='blue',label="auto-correlation signal 1")
plt.plot(ttc,c22,color='green',label="auto-correlation signal 2")
plt.xlabel('$\Delta t$')
plt.ylabel('$signal$')
plt.grid(True)
plt.legend()
plt.title('Correlations')
plt.show()
```
