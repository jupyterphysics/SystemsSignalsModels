##### Initialisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

##### Set the parameter values for this problem

This problem is to illustrate the effect of (non)-matching of the number of samples in relation to the frequency of the signal and the sample frequency. In case this number do not match, this becomes clearly visible in the Fourier specturm.
Play with f0, to experience the effect. 

```python
nn=100   # number of points in the time signal
fs=100.  # sample frequency
f0=10.5  # frequency of the time signal
```

Define the time signal

```python
x = np.linspace(0,nn-1,nn,dtype=complex)
for i in range(nn):
    x[i] = np.cos(2 * np.pi * f0 *i/fs)
```

Plot the function

```python
plt1=plt.figure()
plt.stem(np.real(x),use_line_collection=True)
plt.xlabel('time')
plt.ylabel('signal')
plt.title('Time signal')
plt.show()
```

Calcualte the Fourier transform of the signal

```python
fx = np.fft.fft(x)
```

```python
plt2=plt.figure()
plt.stem(np.abs(fx),use_line_collection=True)
plt.xlabel(r'$\omega$')
plt.ylabel('fft(x)')
plt.grid(True)
plt.title('FFT of time signal')
plt.show()
```
