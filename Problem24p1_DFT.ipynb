Import the essestial libraries

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread("fig/DFTnumpyDef.png")
plt.figure(figsize=(12, 12))
plt.imshow(image)
plt.axis('off')
plt.show()
```

Define the signals

```python
s1 = np.array([1, -1, 1, -1, 1, -1, 1, -1],dtype=float)
s2 = np.array([1, -1, 1, -1, 1, -1, 1],dtype=float)
```

```python
dfts1 = np.fft.fft(s1)
dfts2 = np.fft.fft(s2)
```

```python
plt.stem(np.real(dfts1),markerfmt='C3o',label='real')
plt.stem(np.imag(dfts1),markerfmt='C0o',label='imag')
plt.xlabel('sample number')
plt.ylabel('Fourier Coeffficient')
plt.grid('True')
plt.legend()
plt.title('DFT of a matched sequence')
```

```python
plt.stem(np.real(dfts2),markerfmt='C3o',label='real')
plt.stem(np.imag(dfts2),markerfmt='C0o',label='imag')
plt.xlabel('sample number')
plt.ylabel('Fourier Coeffficient')
plt.grid('True')
plt.legend()
plt.title('DFT of a non-matched sequence')
```
