# Problem 19.3

This problem illustrates the calculation of Probability Density Function (PDF) of time series and the calculation of statistical moments of the distributions and their more suitable equivalences.

Import essential packages

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Read data from files

```python
f2=np.loadtxt("data/signal2LevelPDF.csv",float)
fg=np.loadtxt("data/signalGaussPDF.csv",float)
```

#### Plot the time-series

```python
plt.plot(fg,color='blue',label='normal-PDF')
plt.plot(f2,color='red',label='two level-PDF')
plt.xlabel('$t [ms]$')
plt.ylabel('$x(t)$')
plt.grid(True)
plt.legend()
plt.title('time-series')
plt.show()
```

#### Make a histogram and plot the results

```python
plt.hist(fg,bins=20,color='blue',label='normal-PDF')
plt.hist(f2,bins=20,color='red',label='two level-PDF')
plt.xlabel('$bins$')
plt.ylabel('$count$')
plt.grid(True)
plt.legend()
plt.title('PDF of the two signals')
plt.show()
```

#### Calculate the statistical moments of the two time-series

```python
momentf2 = np.zeros(5, dtype=float)
momentfg = np.zeros(5, dtype=float)

for i in range(momentf2.size):
    momentf2[i] = stats.moment(f2,moment=i)
for i in range(momentfg.size):
    momentfg[i] = stats.moment(fg,moment=i)

print('\nThe first 5 statistical moments of the 2-level-PDF are (starting from zero): ')
print(momentf2)
print('\nThe first 5 statistical moments of the Normal-dist PDF are (starting from zero): ')
print(momentfg)
```

#### Present the statistical moments in more suitable representation

```python
meanf2 = np.mean(f2)
stdevf2 = np.std(f2)
varf2 = np.var(f2)
moment3f2 = stats.moment(f2,moment=3)
kurtf2 = stats.kurtosis(f2)
skewf2 = stats.skew(f2)

meanfg = np.mean(fg)
stdevfg = np.std(fg)
varfg = np.var(fg)
moment3fg = stats.moment(fg,moment=3)
kurtfg = stats.kurtosis(fg)
skewfg = stats.skew(fg)

print('values for the 2-level-PDF: mean = {:.3f}, var = {:.3f},stdev = {:.3f}, skewness = {:.3f}, kurtosis = {:.3f}\n'.format(meanf2,varf2,stdevf2,skewf2,kurtf2))
print('values for the Normal-PDF : mean = {:.3f}, var = {:.3f},stdev = {:.3f}, skewness = {:.3f}, kurtosis = {:.3f}\n'.format(meanfg,varfg,stdevfg,skewfg,kurtfg))

```
