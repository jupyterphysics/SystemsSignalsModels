##### Initalisation

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

#### Transfer functions of the model

```python
k=1
nsys = np.array([k])
dsys = np.array([0.16, 0.285, 1, 1])
sys_tf1 = sig.TransferFunction(nsys,dsys)
# sys_tf1 = sig.TransferFunction([k],[0.16, 0.285, 1, 1])
sys_zpk1 = sys_tf1.to_zpk()
sys_zpk1
```

Make a Nyquist-plot of the function

```python
w,h = sig.freqresp(sys_zpk1)
plt.figure()
plt.plot(h.real, h.imag, 'b')
plt.plot(h.real, -h.imag, 'r')
plt.xlabel(r'Re$(H(\omega))$')
plt.ylabel(r'Im$(H(\omega))$')
plt.grid(True)
plt.show()
```
