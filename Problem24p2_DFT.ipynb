##### Initialisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

Set the parameters for this problem

```python
nn=16  # set the number of points
k=nn-2 # set the k-value
```

Define the time array

```python
x = np.linspace(0,nn-1,nn,dtype=complex)
for i in range(nn):
    x[i]=np.exp((0+1j)*2 *np.pi*k*i/nn)
```

Plot the function values 

```python
fig, plt1 = plt.subplots(1,2,figsize=(9, 3))
plt1[0].stem(np.real(x),use_line_collection=True)
plt1[0].set_xlabel('time')
plt1[0].set_ylabel('Re(x)')
plt1[0].set_title('Real part')
#
plt1[1].stem(np.imag(x),use_line_collection=True)
plt1[1].set_xlabel('time')
plt1[1].set_ylabel('Im(x)')
plt1[1].set_title('Imaginary part')
#
# fig.suptitle(r'Real and Imaginary part of the signal')
plt.show()
```

```python
fx=np.fft.fft(x)
```

```python
plt3=plt.figure()
plt.stem(np.abs(fx),use_line_collection=True)
plt.xlabel(r'$\omega$')
plt.ylabel('Re(FFT(x)')
plt.grid(True)
plt.title('Fourier Transform of the Signal')
plt.show()
```
