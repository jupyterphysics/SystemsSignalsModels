# Example 5.1 - First-order differential equation

##### Initialisation

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

#### Defintition of the model

Create a function that holds the model

```python
def model(y,t,k):
    dydt = -k * y
    return dydt
```

Initial condition

```python
y0 = 5
```

Define the time-axis

```python
t = np.linspace(0,20)
```

Solve the differential equation for various values of the parameter $k$.

```python
k = 0.1
y1 = odeint(model,y0,t,args=(k,))
k = 0.2
y2 = odeint(model,y0,t,args=(k,))
k = 0.5
y3 = odeint(model,y0,t,args=(k,))
```

Plot the results

```python
plt.plot(t,y1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,y3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.title('Solutions of a first-order diff. eqn.')
plt.show()
```
