```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

##### Signals (dc and ac)

```python
dc = np.array([1,1,1,1,1,1])
ac = np.array([1, -1, 1, -1, 1, -1])
```

##### Filters (low-pass -- high-pass)

```python
lp = np.array([1, 1])
hp = np.array([1, -1])
```

##### Apply both filters on each of the signals

```python
dclp = sig.convolve(dc,lp)
dchp = sig.convolve(dc,hp)
aclp = sig.convolve(ac,lp)
achp = sig.convolve(ac,hp)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(221)
plt.stem(dclp,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('signal')
plt.title('DC signal through low-pass filter')
plt.grid(True)
plt.subplot(222)
plt.stem(dchp,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('signal')
plt.title('DC signal through high-pass filter')
plt.grid(True)
plt.subplot(223)
plt.stem(aclp,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('signal')
plt.title('AC signal through low-pass filter')
plt.grid(True)
plt.subplot(224)
plt.stem(achp,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('signal')
plt.title('AC signal through high-pass filter')
plt.grid(True)
plt.show()
```
